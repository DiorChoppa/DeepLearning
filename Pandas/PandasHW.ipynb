```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('sample_data/BlackFriday.csv')
df
```

```python
#Сколько всего возрастных категорий

print(len(df['Age'].unique()))

df['Age'].unique()
```

```python
#Сколько строк с мужчинами из категории города A

print(len(df[(df['City_Category'] == 'A') & (df['Gender'] == 'M')]))

df[(df['City_Category'] == 'A') & (df['Gender'] == 'M')]
```

```python
#Сколько женщин от 46 до 50, потративших (столбец Purchase) больше 20000

print(len(df[(df['Gender'] == 'F') & (df['Age'] == '46-50') & (df['Purchase'] > 20000)]))

df[(df['Gender'] == 'F') & (df['Age'] == '46-50') & (df['Purchase'] > 20000)]
```

```python
#Сколько NaN'ов в столбце Product_Category_3?

df['Product_Category_3'].isna().sum()
```

```python
#Какую долю (округлить до 4-го знака) от всех покупателей составляют мужчины 26-35 лет и женщины > 36 лет

male = len(df[(df['Gender'] == 'M') & (df['Age'] == '26-35')])
female = len(df[(df['Gender'] == 'F') & (df['Age'] != '0-17') & (df['Age'] != '18-25') & (df['Age'] != '26-35')])
print(round((male+female)/len(df), 4))
```
